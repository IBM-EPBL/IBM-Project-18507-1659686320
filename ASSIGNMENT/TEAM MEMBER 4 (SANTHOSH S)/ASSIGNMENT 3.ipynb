# Check for Categorical Columns and perform encoding

```python
df.info()
```

```python
df['Sex'].unique()
```

```python
df=pd.get_dummies(df,columns=['Sex'])
df.head()
```


# Split the data into dependent and independent variables

```python
y=df['Rings'].values                           #dependent variables
y
```

```python
x=df.drop(columns=['Rings'],axis=1).values     #independent variables
x
```

# Split the data into training and testing

```python
x_train,x_test,y_train,y_test = train_test_split(x,y,test_size=0.2,random_state=0)
```

# Build the model

```python
lr= LinearRegression()
```

```python
lr.fit(x_train,y_train)          #Train the model
```

# Test the model

```python
y_prediction= lr.predict(x_test)
y_prediction
```

# Measure the performance using Metrics

```python
#score < 1.0 , means that all predictions matches the expected values exactly.
score = r2_score(y_test,y_prediction)
score
```

```python
#error value < 1.0 , the prediction succeeds
MSE = mean_absolute_percentage_error(y_test,y_prediction)
MSE
```

```python
# distance < 1.0, prediction is accurate
MPD = mean_poisson_deviance(y_test,y_prediction)
MPD
```
